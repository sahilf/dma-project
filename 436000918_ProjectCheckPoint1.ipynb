#Data quality issues

*   Amazon Fine Food Reviews : Since these are web scraped, the text is often cluttered with html tags, unescaped special characters, and emojis that can interfere with standard tokenizers. Reviews where the same user posts identical text for different sizes or flavors of the same product, which can artificially inflate word frequency counts.

*   Tweets about Top Companies Stock : Presence of Re-tweets and copy-pasted spam tweets from bot accounts can significantly skew the dataset. Text is  sparse, filled with shortened URLs, ticker symbols and hashtags, making it difficult to extract traditional sentence structures.


*   Sarcasm on Reddit : The biggest quality issue is that the child comment (labeled as sarcastic) may be semantically meaningless without the parent_comment, making independent text mining of the labels difficult. Also, presence of Reddit-specific formatting (like > quotes, bolding, or table structures) can create dirty tokens if the text isn't properly stripped during the cleaning phase.

#Algorithmic feasibility

*   Amazon Fine Food Reviews : Text mining on this dataset is highly feasible due to the structured nature of the reviews, which allows for efficient vectorization using TF-IDF or Bag-of-Words across the 500k entries. Since the corpus contains significant noise like HTML tags and product variations, the primary computational overhead lies in the preprocessing and deduplication stages rather than the mining algorithms themselves. From a graph mining perspective, constructing a bipartite user-product graph is computationally manageable on a standard machine using sparse matrix representations to identify central product hubs.

*   Tweets about Top Companies Stock : The high velocity and short form nature tweets require memory efficient streaming or batch processing to maintain feasibility during text mining operations. Because Twitter data is extremely sparse and contains non-standard language, frequency-based mining and N-gram analysis are the most efficient ways to identify market-moving keywords without excessive CPU load. Also, implementing burst detection algorithms is feasible to pinpoint sudden spikes in tweet volume that correlate with significant financial market events.


*   Sarcasm on Reddit : Detecting sarcasm remains a complex linguistic challenge, but the balanced nature of this dataset makes supervised mining tasks statistically stable and algorithmically feasible. Feature engineering focusing on punctuation density and sentiment-reversal patterns between parent and child comments can be processed efficiently without requiring massive GPU clusters.

#Bias considerations

*   Amazon Fine Food Reviews : There may be a significant rating distribution bias, where the data is heavily skewed toward 5-star ratings, potentially making it difficult to get distinct patterns for mediocre products

*   Tweets about Top Companies Stock : This dataset may contain a high degree of bot driven bias, where automated accounts repeatedly tweet specific ticker symbols to artificially inflate sentiment or create a false sense of hype.

*   Sarcasm on Reddit : The data might exhibit community specific bias, where certain subreddits have their own inside jokes or linguistic norms that are interpreted as sarcasm in one context but not in another.

#Ethical considerations

*   Amazon Fine Food Reviews : Automated sentiment models can be used to unfairly suppress valid negative feedback if companies use these tools to identify and flag reviews that hurt their brand, potentially misleading future customers.

*   Tweets about Top Companies Stock :The main ethical risk is the amplification of market manipulation. Mining hype without identifying bot driven sentiment could lead to the spread of misinformation.

*   Sarcasm on Reddit : Involves the risk of misinterpretation, where sarcastic remarks could be flagged as genuine harassment or hate speech by automated systems, leading to unfair moderation. Because sarcasm is deeply contextual, there is also a demographic bias concern, as mining models may fail to account for how different regional or marginalized groups use irony.

#(D) Exploratory Data Analysis



````python
#Code

import pandas as pd
from google.colab import drive


drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/TAMU/TAMU Study/IV_Sem/Data Mining/Project/Reviews.csv'

df = pd.read_csv(file_path)

rows, columns = df.shape
print(f"\nNumber of rows: {rows}")
print(f"\nNumber of columns: {columns}")
print("\n")
print(df.info())
df.head()
````

````python
#Code

import matplotlib.pyplot as plt
import seaborn as sns

#sort by ProductId to maintain a consistent first instance of a review
df_sorted = df.sort_values('ProductId', axis=0, ascending=True)

#drop duplicates where the User, ProfileName, and Text are identical.
df_cleaned = df_sorted.drop_duplicates(subset={"UserId","ProfileName","Text"}, keep='first', inplace=False)

# helpfulness sanity check
df_cleaned = df_cleaned[df_cleaned.HelpfulnessNumerator <= df_cleaned.HelpfulnessDenominator]

# date & time conversion
df_cleaned['DateTime'] = pd.to_datetime(df_cleaned['Time'], unit='s')
df_cleaned['Year'] = df_cleaned['DateTime'].dt.year
df_cleaned['Month'] = df_cleaned['DateTime'].dt.month

# text length analysis
df_cleaned['Text_Word_Count'] = df_cleaned['Text'].apply(lambda x: len(str(x).split()))


print(f"\n Original Rows: {len(df)}")
print(f"\n Post Data Cleaning Rows: {len(df_cleaned)}")
print(f"\n Data spans from: {df_cleaned['DateTime'].min().year} to {df_cleaned['DateTime'].max().year}")
print(f"\n Unique Users: {df_cleaned['UserId'].nunique()}")
print(f"\n Unique Products: {df_cleaned['ProductId'].nunique()}")


plt.figure(figsize=(10,5))
df_cleaned.groupby('Year')['Id'].count().plot(kind='line', marker='o')
plt.title('Chronological Distribution of Reviews')
plt.ylabel('Review Count')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````

````python
#Code

from collections import Counter
import re

def clean_html(text):
    # Remove HTML tag. This regex looks for anything between < and >
    clean_text = re.sub(r'<.*?>', ' ', str(text))

    #replace them with their actual characters
    clean_text = clean_text.replace('&quot;', '"').replace('&amp;', '&').replace('&nbsp;', ' ')

    #Remove extra whitespace created by the stripping
    clean_text = " ".join(clean_text.split())

    return clean_text

stop_words = set(['the', 'and', 'a', 'to', 'of', 'is', 'it', 'for', 'in', 'i', 'this', 'was', 'with', 'my', 'but', 'as', 'are', 'not', 'have', 'that'])

def get_top_bigrams(text_series, n=10):
    all_bigrams = []
    for text in text_series:
        #convert to lowercase and remove non alphabetic chars
        words = re.findall(r'\b\w+\b', str(text).lower())
        #remove stop words
        words = [w for w in words if w not in stop_words]
        # create pairs (bigrams)
        bigrams = zip(words, words[1:])
        all_bigrams.extend([" ".join(bi) for bi in bigrams])

    return Counter(all_bigrams).most_common(n)

df_cleaned['Text'] = df_cleaned['Text'].apply(clean_html)
df_cleaned['Summary'] = df_cleaned['Summary'].apply(clean_html)

assert clean_html("Hello<br />World") == "Hello World", "Failed to remove <br />"
assert clean_html("<b>Bold Text</b>") == "Bold Text", "Failed to remove <b> tags"
assert clean_html("Salt &amp; Pepper") == "Salt & Pepper", "Failed to convert &amp;"
assert clean_html("He said &quot;Hi&quot;") == 'He said "Hi"', "Failed to convert &quot;"
assert clean_html("<div>  Content  <p>More</p>  </div>") == "Content More", "Failed to handle nested tags or whitespace"

print("\nTest Cases Passed")

#Get 1-Star and 5-Star
text_1star = df_cleaned[df_cleaned['Score'] == 1]['Text']
text_5star = df_cleaned[df_cleaned['Score'] == 5]['Text']


print("\nTop 10 Bi-grams for 1-Star Reviews:")
print(get_top_bigrams(text_1star,20))

print("\nTop 10 Bi-grams for 5-Star Reviews:")
print(get_top_bigrams(text_5star))
````

````python
#Code

sns.set_style("whitegrid")

# Frequency of each Review Score
plt.figure(figsize=(10, 6))
sns.countplot(x='Score', data=df_cleaned, palette='viridis')
plt.title('Total Number of Reviews per Score', fontsize=15)
plt.xlabel('Review Score (1-5)', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.show()

# Average Word Count per Review Score
avg_word_count = df_cleaned.groupby('Score')['Text_Word_Count'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Text_Word_Count', data=avg_word_count, palette='magma')
plt.title('Average Word Count by Review Score', fontsize=15)
plt.xlabel('Review Score (1-5)', fontsize=12)
plt.ylabel('Average Number of Words', fontsize=12)
plt.show()

# Average Helpfulness Ratio per Word Count Bin
df_help = df_cleaned[df_cleaned['HelpfulnessDenominator'] > 0].copy()
df_help['Helpfulness_Ratio'] = df_help['HelpfulnessNumerator'] / df_help['HelpfulnessDenominator']

# Binning word counts
df_help['Word_Count_Bin'] = pd.cut(df_help['Text_Word_Count'],
                                  bins=[0, 20, 50, 100, 200, 500, 1000],
                                  labels=['0-20', '21-50', '51-100', '101-200', '201-500', '500+'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Word_Count_Bin', y='Helpfulness_Ratio', data=df_help, palette='coolwarm')
plt.title('Average Helpfulness Ratio vs. Review Length', fontsize=15)
plt.xlabel('Review Word Count Range', fontsize=12)
plt.ylabel('Average Helpfulness Ratio', fontsize=12)
plt.ylim(0, 1)
plt.show()
````

````python
#Code

#Review Count per User
user_counts = df_cleaned.groupby('UserId')['Id'].count().reset_index()
user_counts.columns = ['UserId', 'ReviewCount']

def segment_user(count):
    if count <= 2:
        return 'Amateur Reviewer (1-2)'
    elif count <= 5:
        return 'Active Reviewer (3-5)'
    else:
        return 'Very Active Reviewer (5+)'

user_counts['User_Segment'] = user_counts['ReviewCount'].apply(segment_user)


plt.figure(figsize=(10, 6))
order = ['Amateur Reviewer (1-2)', 'Active Reviewer (3-5)', 'Very Active Reviewer (5+)']
sns.countplot(x='User_Segment', data=user_counts, order=order, palette='Set2')

plt.title('User Segmentation', fontsize=15)
plt.xlabel('User Segment', fontsize=12)
plt.ylabel('Number of Unique Users', fontsize=12)
plt.show()

print(user_counts['User_Segment'].value_counts())
````

Starting with 568,454 original records, the volume was reduced to 393,645 unique rows after deduplication and noise removal. This significant reduction occurred because many reviews were repeated across different product variations, such as different sizes or flavors of the same item. Stripping out noise like HTML tags ensured that the focus remained on human language, preventing website code from interfering with the linguistic results.

The cleaned data spans a long timeline from 1999 to 2012, capturing over a decade of consumer behavior. With 256,059 unique users and 67,488 unique products, the scale of the dataset is large enough to show meaningful trends. The data shows a classic positivity bias, with a massive majority of reviews being 5 stars. Interestingly, even though 5-star reviews are more common, the 1-star reviews tend to be longer on average. This suggests that dissatisfied customers provide more descriptive feedback as they attempt to justify their complaints. This length was also found to be linked to community perception. Reviews with higher word counts generally receive more helpful votes from other users, confirming that detailed content is viewed as more authoritative.

The users themselves were categorized to prepare for more advanced networking analysis. By grouping the unique users into categories like Amateurs, Active, and Very Active based on how many reviews they submitted, could identify a small but influential group of frequent reviewers. These frequent reviewers act as the primary connectors in the dataset, linking many different products together through their buying habits.

Also, examined two-word phrases, known as bi-grams from the text reviews. This provided a much clearer picture of the actual context behind the reviews. By isolating the most common phrases in top-rated versus bottom-rated reviews, distinct patterns emerged, such as "highly recommend" for positive experiences and "very disappointed" for negative ones. This linguistic profiling creates a strong foundation for identifying anomalies, as it makes it easy to spot instances where the written text does not logically match the numerical score provided by the user.
